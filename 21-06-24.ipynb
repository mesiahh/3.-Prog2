# Sorting Algorithms

**Bubble Sort** - a simple comparison-based algorithm where each pair of adjacent elements is compared and the elements are swapped if they are in the wrong order. This process repeats until the list is sorted.

Bubble Sort Example:

```python
%% file sortbubble.c

void bubbleSort(int* arr, int size) {
    int i, j, temp;
    for (i = 0; i < size - 1; i++) {
        for (j = 0; j < size - i - 1; j++) {
            if (arr[j] > arr[j + 1]) {
                // Swap arr[j] and arr[j + 1]
                temp = arr[j];
                arr[j] = arr[j + 1];
                arr[j + 1] = temp;
            }
        }
    }
}
```

<br>

**Insertion Sort** - builds the final sorted array one item at a time. It takes each element from the input data and finds the location it belongs within the sorted list and inserts it there.

Insertion Sort Example:


```python
%% file sortbubble.c

void insertionSort(int* arr, int size) {
    int i, key, j;
    for (i = 1; i < size; i++) {
        key = arr[i];
        j = i - 1;
        // Move elements of arr[0..i-1], that are greater than key, to one position ahead of their current position
        while (j >= 0 && arr[j] > key) {
            arr[j + 1] = arr[j];
            j = j - 1;
        }
        arr[j + 1] = key;
    }
}
```

**Structures in C**

A structure in C is a user-defined data type that groups related variables of different data types into a single unit.

Defining and Using a Structure:

```python
#include <stdio.h>

// Define a structure named 'Person'
struct Person {
    char name[50];
    int age;
    float height;
};

int main() {
    // Declare a variable of type 'Person'
    struct Person person1;

    // Access and assign values to the structure members
    printf("Enter name: ");
    scanf("%s", person1.name);
    printf("Enter age: ");
    scanf("%d", &person1.age);
    printf("Enter height: ");
    scanf("%f", &person1.height);

    // Print the values
    printf("\nName: %s\nAge: %d\nHeight: %.2f\n", person1.name, person1.age, person1.height);

    return 0;
}
```

<br>

**Accessing Structure Members:**

Dot Operator (.): Used to access members of a structure variable.
Arrow Operator (->): Used to access members of a structure through a pointer to the structure.

Example of Accessing Structure Members:

```python
%% file acccessstrucmem.c

#include <stdio.h>

// Define a structure named 'Person'
struct Person {
    char name[50];
    int age;
    float height;
};

int main() {
    struct Person person1;
    struct Person* ptr = &person1;

    // Access members using dot operator
    person1.age = 30;
    printf("Age using dot operator: %d\n", person1.age);

    // Access members using arrow operator
    ptr->height = 5.9;
    printf("Height using arrow operator: %.2f\n", ptr->height);

    return 0;
}
```
